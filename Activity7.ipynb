```python
import folium
from folium.plugins import HeatMap
import numpy as np
import pandas as pd
import warnings 
warnings.filterwarnings('ignore')
```

<h1>Heatmap Data from Global Terrorism</h1>
<h2>Analyst: Ian Genesis D. Minorca</h2>

```python
kaboom = pd.read_csv('Global_terrorist.csv', encoding='ISO-8859-1')
kaboom
```

```python
fil1 = kaboom[kaboom['region_txt'] == 'South America']
fil2 = fil1[fil1['iyear'] >= 2000]
fil3 = fil2[fil2['attacktype1_txt'] == 'Bombing/Explosion']
fil4 = fil3[fil3['nkill'] >= 10]
fil4
```

```python
spatial = fil4.groupby(['longitude', 'latitude']).size().reset_index(name='terror_count')
spatial
```

<h1>HeatMap 1</h1>
<h3>Bomb Explosion incidents that happened in South America</h3>

```python
agg_df = fil1.groupby(['longitude', 'latitude']).size().reset_index(name='incident_count')

init_mapa = agg_df[['latitude', 'longitude', 'incident_count']].values.tolist()
mapa = folium.Map(location=[-15, -60], zoom_start = 4)
HeatMap(init_mapa).add_to(mapa)
mapa.save('terrorism_sa.html')
mapa
```

<h1>HeatMap 2</h1>
<h3>Assination records in Central America & Caribbean</h3>

```python
ca1 = kaboom[kaboom['region_txt'] == 'Central America & Caribbean']
ca2 = ca1[ca1['iyear'] >= 2000]
ca3 = ca2[ca2['attacktype1_txt'] == 'Assassination']

agg_df2 = ca3.groupby(['longitude', 'latitude']).size().reset_index(name='incident_count')

init_mapa_ca = agg_df2[['latitude', 'longitude', 'incident_count']].values.tolist()
mapa2 = folium.Map(location=[14, -91], zoom_start = 4)
HeatMap(init_mapa_ca).add_to(mapa2)
mapa2.save('terrorism_ca.html')
mapa2
```
